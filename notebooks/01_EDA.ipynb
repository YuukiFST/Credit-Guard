# 01 — Exploratory Data Analysis (EDA)

Este notebook realiza a exploração inicial do dataset UCI Credit Card, focando em:
1. Distribuição da variável target (desbalanceamento)
2. Correlações e impacto das features
3. Análise de viés/fairness (LGPD)

```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
from src.data.loader import load_raw_data
from src.models.fairness import calculate_fairness_metrics

pd.options.display.max_columns = None
```

## 1. Carregamento e Visão Geral

```python
df = load_raw_data()
df.head()
```

## 2. Análise da Variável Target (Inadimplência)

```python
target_counts = df['default.payment.next.month'].value_index_labels = {0: 'Adimplente', 1: 'Inadimplente'}
fig = px.pie(
    df, 
    names='default.payment.next.month', 
    title='Distribuição de Inadimplência (Target)',
    labels={'default.payment.next.month': 'Inadimplente'}
)
fig.show()
```

## 3. Análise de Viés (Fairness)

Analisando se existe disparidade na taxa de inadimplência por sexo (SEX) e educação (EDUCATION).

```python
fig = px.histogram(
    df, 
    x='SEX', 
    color='default.payment.next.month', 
    barmode='group',
    title='Inadimplência por Sexo (1=M, 2=F)'
)
fig.show()
```

## 4. Correlação

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
plt.title('Mapa de Correlação de Pearson')
plt.show()
```
